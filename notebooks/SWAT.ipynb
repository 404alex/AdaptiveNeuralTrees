## (i) Train an ANT on the IoT image recognition dataset
From the code directory, run the following command to train an ANT:

```bash
python tree.py --dataset swat --experiment demo4 --subexperiment swat_ant --batch-size 256 \
--epochs_patience 5 --epochs_node 100 --epochs_finetune 200 -t_ver 3 -t_ngf 5 -t_k 3 -r_ver 1 \
-r_ngf 5 -r_k 3 -s_ver 1 -ds_int 1 --maxdepth 10 --num_workers 0 --seed 1
```

It takes less than 3 hours on a single Titan X GPU.
